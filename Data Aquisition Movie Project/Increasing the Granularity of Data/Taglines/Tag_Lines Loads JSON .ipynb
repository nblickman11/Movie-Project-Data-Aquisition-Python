Below is the output of `Counter` object after applying it over 100 movies, loaded from the previously-created JSON file.

This dictionary will help us count the number of words in each tagline (Taglines themselves were printed first).

```python
from collections import Counter

import json
file = open("tag_lines.json", "r").read()
tag_lines = json.loads(file)
final = {}
c = 1
for key, t_lines in tag_lines.items():
    print("Movie Number:", key)
    for line in t_lines:
        print(line.strip())
    movie_count = Counter()
    for t_line in t_lines:
        t_line = t_line.strip()
        t_line = t_line.split(" ")
        for word in t_line:
            movie_count[word] += 1
    final.update({str(c):movie_count})
    c += 1
    print()

print("\n\n\n\n")
print("DISPLAY ABOVE")
print("WORD COUNT BELOW")
print("\n\n\n\n")

for key, values in final.items():
    print("Movie Number:", key)
    print(values)
    print()
```
