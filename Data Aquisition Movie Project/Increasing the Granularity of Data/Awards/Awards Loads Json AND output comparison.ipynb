Below code is the same as what he had in the "Awards Main Code", however we pulled it over to run the outputs and it is a different structure than the Counters above.
    
- Some of the advantages of using the two dictionaries, by separating Winners and Nominees (as seen in the output below), is that it separates Awards and Nominees, the details of each award are in order, and it contains the individual names of people that won individual awards. 


- The Counter Structure does not have those advantages. In fact, it does face a couple of limitations.  As mentioned, it does not list the name of a winner of an individual award, and it may not capture the very few awards that don't start with "Best". 

```python
i = 1
for item in range(1,2):
    url = 'https://www.imdb.com/list/ls003495084/?sort=date_added,desc&st_dt=&mode=detail&page=' + str(item)
    html_text = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html_text, 'html.parser')
    movies={}
    for x in soup.find_all('a'):
        if x.get("href") is not None:
            if re.search('^/title/tt', x['href']) and x.text!=' \n':
                movies[x.text]={'link':'https://www.imdb.com/'+x['href']}     
    
    movie_numbers = []
    for key, dictionary in movies.items():
        for values in dictionary.values():
            movie_number = re.search("[0-9]+", values)
            movie_numbers.append(movie_number.group(0))

    Awards = {}
    for number in movie_numbers:
        def get_url(M):
            return urllib.request.urlopen("https://www.imdb.com/title/tt" + M + "/awards?ref_=tt_ql_op_1").read()
        html_text = get_url(number)    
        movie = BeautifulSoup(html_text, 'html.parser')
        
        tds = movie.find_all("td")
        d = defaultdict(list)
        Win_Nom = ''
        
        try:
            for td in tds:
                split_line = (td.text.split("\n"))
                for line in split_line:
                    if line == "Nominee":
                        Win_Nom = 'Nominee'
                    elif line == "Winner":
                        Win_Nom = "Winner"
                    elif Win_Nom == 'Nominee':
                        piece = (line.split("  "))
                        if piece[0] != '': 
                            d["Nominee"].append(line)    
                        if len(piece) > 5:
                            piece[6].strip()
                            d["Nominee"].append(piece[6])
                    elif Win_Nom == "Winner":
                        piece = line.split("  ")
                        if piece[0] != '':
                            d["Winner"].append(line)
                        if len(piece) > 1:
                            piece[6].strip()
                            d["Winner"].append(piece[6])
                            
        except IndexError:
            for td in tds:
                split_line = (td.text.split("\n"))
                for line in split_line:
                    if line == "Nominee":
                        Win_Nom = 'Nominee'
                    elif line == "Winner":
                        Win_Nom = "Winner"
                    elif Win_Nom == 'Nominee':
                        piece = (line.split("  "))
                        d["Nominee"].append(line)    
                    elif Win_Nom == "Winner":
                        piece = line.split("  ")
                        d["Winner"].append(line)      

        for keys, values in d.items():
            print(keys)
            print(values)
```
