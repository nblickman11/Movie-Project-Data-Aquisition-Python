```python
from collections import Counter

import json
file = open("plot_summary.json", "r").read()
summaries = json.loads(file)
#for key, value in summaries.items():
    #print(key)
    #print(value)
    #print()
    #print("Next Movie:")  

    
    
# Creating Word Counts for the Synopsis of every movie.     
final = {}
i = 1
for key, values in summaries.items():
    lst = values.split(" ")
    movie_count = Counter()
    for item in lst:
        movie_count[item] += 1
    final.update({str(i):movie_count})
    i+=1

for key, values in final.items():
    print(key)
    print(values)
    print()
    print("Next Movie:") 
  
```
